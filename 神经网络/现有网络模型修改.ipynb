# 网络模型的保存与读取

```python
import torch
import torchvision
vgg16 =torchvision.models.vgg16(weights=False)
#重建结构只需要 1 秒钟，但训练可能需要几天。 这就是保存的意义
#保存方式1，同时保存网络模型结构和数据
torch.save(vgg16,"vgg16_method1.pth")
#保存方式2,训练数据保存为字典不保存网络模型结构(用的空间小)
torch.save(vgg16.state_dict(),"vgg16_method2.pth")


#读取模型
model1 = torch.load("vgg16_method1.pth",weights_only=False)


#   model2 = torch.load("vgg16_method2.pth",weights_only=False)      #因为保存为了字典所以还要恢复网络模型结构
vgg16.load_state_dict(torch.load("vgg16_method2.pth",weights_only=False))

# print(model1)
# print(model2)
```
