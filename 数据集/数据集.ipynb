# 数据集代码写法

```python
from torch.utils.data import Dataset
import cv2
import os #为了获取图片的地址

class MyData(Dataset):
    
    def __init__(self,root_dir,label_dir):#每个文件的当前路径和他的标签（标签刚好就是上一级的路径名）
         #pass函数体不能是空的要用pass占位
        self.root_dir = root_dir     #把传入的参数 root_dir 保存到实例变量 self.root_dir 中，让这个类的所有方法都能访问到它。
        self.label_dir = label_dir
        self.path = os.path.join(root_dir,label_dir)#内部变量而不是参数。
        self.img_path = os.listdir(self.path) # 获取该路径下所有文件名
        
    def __getitem__(self,index):
        img_name = self.img_path[index]  # ✅ 用 self.img_path，其索引指向的就是文件名
        img_item_path = os.path.join(self.root_dir,self.label_dir,img_name)#注意不是self.img_name这是全部文件名而不是上面一行的局部变量
        img = cv2.imread(img_item_path)
        label = self.label_dir
        return img,label,img_name
        
    def __len__(self):
        return len(self.img_path)
        
root_dir ="dataset/train"
ants_label_dir = "ants"
bees_label_dir = "bees"
ants_dataset = MyData(root_dir,ants_label_dir)
bees_dataset = MyData(root_dir,bees_label_dir)

img, label, img_name = ants_dataset[4]
print(f"图片形状: {img.shape}")#输出该图片的高，宽，通道数
print(f"文件名: {img_name}")
print(f"标签: {label}")

img, label, img_name = bees_dataset[4]
print(f"图片形状: {img.shape}")#输出该图片的高，宽，通道数
print(f"文件名: {img_name}")
print(f"标签: {label}")

#数据集可以相加
train_dataset = ants_dataset + bees_dataset  #拼接
print(len(ants_dataset))
print(len(bees_dataset))
print(len(train_dataset))
img, label, img_name = train_dataset[100]
print(f"标签: {label}")
```
