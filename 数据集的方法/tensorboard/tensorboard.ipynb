```python
from  torch.utils.tensorboard import SummaryWriter #SummaryWriter + TensorBoard 能帮你把这些数字变成实时更新的图表
import numpy as np
from PIL import Image
import cv2

# ========== 第一步：创建 SummaryWriter ==========
writer = SummaryWriter("logs")#创建一个实例存储到logs文件夹下



# ========== 第二步：记录各种数据 ==========

img_path ="/Users/haoye/anaconda_projects/basic_learning/pytorch/tensorboard/dataset/train/ants_image/0013035.jpg"


##Image.open() 返回的是 PIL 图像对象。先接收然后转成numpy必须分两步
img_PIL= Image.open(img_path)
img_array = np.array(img_PIL)

##用OpenCV直接一步到位得到numpy矩阵
img_array = cv2.imread(img_path)


print(img_array.shape)  #注意格式，add_image有规定位置的这里打印出来是(h,w,c),所以后面要跟着dataformats=‘hwc’
writer.add_image("test",img_array,1,dataformats='HWC')  #1指的是步数，时间戳



#y=2x
for i in range(100):
    writer.add_scalar("y=2x",i*2,i)#第一个位置是别名，第二个位置是y轴，第三个位置是x轴。如果别名不变但是改了旧的会叠上去



# ========== 第三步：关闭 writer ==========
writer.close()

print("数据已写入 logs 文件夹")
print("在终端运行: tensorboard --logdir=logs")
```
