# ResNet模型

### 网络各层结构

```python
输入: [batch, 3, 32, 32]  (CIFAR-10 版本) batch是一次传进网络的图片数量
                    │
                    ▼
┌─────────────────────────────────────────────────────────────┐
│  conv1: 3→64, 3x3, stride=1, padding=1                      │
│  bn1 + relu                                                  │
│  输出: [batch, 64, 32, 32]                                   │
└─────────────────────────────────────────────────────────────┘
                    │
                    ▼
┌─────────────────────────────────────────────────────────────┐
│  layer1: 2个 BasicBlock，每个保持尺寸不变                    │
│  输出: [batch, 64, 32, 32]                                   │
└─────────────────────────────────────────────────────────────┘
                    │
                    ▼
┌─────────────────────────────────────────────────────────────┐
│  layer2: 2个 BasicBlock，第1个 stride=2 (下采样)             │
│  输出: [batch, 128, 16, 16]                                  │
└─────────────────────────────────────────────────────────────┘
                    │
                    ▼
┌─────────────────────────────────────────────────────────────┐
│  layer3: 2个 BasicBlock，第1个 stride=2 (下采样)             │
│  输出: [batch, 256, 8, 8]                                    │
└─────────────────────────────────────────────────────────────┘
                    │
                    ▼
┌─────────────────────────────────────────────────────────────┐
│  layer4: 2个 BasicBlock，第1个 stride=2 (下采样)             │
│  输出: [batch, 512, 4, 4]                                    │
└─────────────────────────────────────────────────────────────┘
                    │
                    ▼
┌─────────────────────────────────────────────────────────────┐
│  全局平均池化: AvgPool2d(kernel_size=4)                      │
│  输出: [batch, 512, 1, 1]                                    │
└─────────────────────────────────────────────────────────────┘
                    │
                    ▼
┌─────────────────────────────────────────────────────────────┐
│  展平: view(batch, -1)                                      │
│  输出: [batch, 512]                                          │
└─────────────────────────────────────────────────────────────┘
                    │
                    ▼
┌─────────────────────────────────────────────────────────────┐
│  全连接层: Linear(512, 10)                                  │
│  输出: [batch, 10]                                           │
└─────────────────────────────────────────────────────────────┘
```

### 第一层参数为什么这样设置？

```python
原始图片: 32×32
加一圈padding: 变成 34×34
3×3卷积核在34×34上滑动，步长1
输出尺寸 = (34 - 3)/1 + 1 = 32×32

结论：padding=1 保证了输出尺寸和输入一样大。只有尺寸相等才能实现残差连接:输出 = F(x) + x
```

### 二三四层下采样了尺寸减半那还怎么残差连接？
* 主路尺寸减半，捷径也会减半，刚好都能相加了

# 为什么下采样会导致感受野变大？

```python
“为了快速读取全图大概信息做出的牺牲” —— 这正是下采样的本质！
原始地图 (32×32)：每个格子代表 1 米×1 米
你只能看到 3×3 个格子 = 9 平方米

下采样后的地图 (16×16)：把 2×2 格子合并成 1 个格子
每个格子现在代表 2 米×2 米 = 4 平方米

同样看 3×3 个格子：
现在看到的是 6 米×6 米 = 36 平方米！
```

### 为什么每层都是两个残差块？

```python
# ResNet 家族的不同配置，其实是试出来的
ResNet18: [2, 2, 2, 2]  # 每个阶段2个残差块
ResNet34: [3, 4, 6, 3]  # 不同阶段不同数量
ResNet50: [3, 4, 6, 3]  # 但使用 Bottleneck（更深的结构）
ResNet101: [3, 4, 23, 3] # layer3 有 23 个！
ResNet152: [3, 8, 36, 3] # 最深，layer3 有 36 个
每个残差块内部包含了主路跟捷径，并不是一个残差块是主路，一个残差块是捷径。两个残差块流水线连接
太多残差块会导致过拟合。1个残差块表达能力不足准确率低。
```

```python
输入 x: [batch, 64, 32, 32]
           │
           ├─────────────────────────────┐
           │                             │
           ▼                             ▼
      主路 (stride=2)                捷径 (stride=2)
           │                             │
    Conv2d(64→128, 3x3)          Conv2d(64→128, 1x1)
    stride=2, padding=1           stride=2
           │                             │
           ▼                             ▼
      尺寸: 16×16                    尺寸: 16×16
      通道: 128                      通道: 128
           │                             │
           └──────────┬──────────────────┘
                      ▼
                 尺寸匹配！可以直接相加！
                 [batch, 128, 16, 16]
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

# ==================== BasicBlock：残差块的基本单元 ====================
# 这是 ResNet18 和 ResNet34 使用的基础残差块
# 对于 ResNet50/101/152，会使用 Bottleneck 块（1x1-3x3-1x1 结构）
class BasicBlock(nn.Module):
    # expansion 表示输出通道数相对于输入通道数的放大倍数
    # BasicBlock 的输出通道数 = 输入通道数 × expansion
    # 这里 expansion=1，说明输出通道数 = 输入通道数（残差块不改变通道数）
    expansion = 1
    
    def __init__(self, in_planes, planes, stride=1):
        """
        参数：
            in_planes: 输入特征图的通道数
            planes: 当前残差块内部卷积层的输出通道数
            stride: 第一个卷积层的步长（1或2），为2时特征图尺寸减半
        """
        super(BasicBlock, self).__init__()
        
        # ========== 主路径（Main Path）==========
        # 第一个卷积层：3x3卷积
        # - stride=stride：当stride=2时，输出尺寸减半（实现下采样）
        # - padding=1：保持卷积前后尺寸不变（当stride=1时）
        # - bias=False：因为后面有BatchNorm，bias会被BN的β参数吸收，不需要
        self.conv1 = nn.Conv2d(in_planes, planes, kernel_size=3, 
                               stride=stride, padding=1, bias=False)
        # BatchNorm：批量归一化，加速训练、稳定梯度、允许更大学习率
        # 原理：对每个batch的每个通道进行归一化（均值0，方差1），再学习缩放γ和平移β
        self.bn1 = nn.BatchNorm2d(planes)
        
        # 第二个卷积层：3x3卷积
        # - stride=1：保持尺寸不变（下采样已经在第一个卷积层完成）
        # - padding=1：保持尺寸不变
        self.conv2 = nn.Conv2d(planes, planes, kernel_size=3, 
                               stride=1, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(planes)
        
        # ========== 捷径连接（Shortcut / Identity Mapping）==========
        # 残差的核心思想：输出 = F(x) + x（主路输出 + 输入）
        # 但需要保证 F(x) 和 x 的维度（通道数、高、宽）一致才能相加
        
        # 默认情况下，shortcut 是恒等映射（直接返回输入 x）
        ## 就是给 shortcut 一个"空壳子"，默认情况下它什么都不做，直接返回输入 x；当需要调整维度时，再把这个空壳子替换成真正的卷积层
        self.shortcut = nn.Sequential()
        
        # 当需要调整维度时（两种情况）：
        # 1. stride != 1：输入和输出的高/宽不一致（需要下采样）
        # 2. in_planes != expansion * planes：输入输出通道数不一致
        if stride != 1 or in_planes != self.expansion * planes:
            self.shortcut = nn.Sequential(
                # 1x1卷积的作用：
                # - 当stride=2时：将高/宽减半（下采样）
                # - 调整通道数：从 in_planes 变为 expansion * planes
                # - 1x1卷积不改变空间结构，只改变通道数
                nn.Conv2d(in_planes, self.expansion * planes, 
                         kernel_size=1, stride=stride, bias=False),
                nn.BatchNorm2d(self.expansion * planes)
            )
    
    def forward(self, x):
        """
        前向传播：
        输出 = ReLU( BN( Conv2( ReLU( BN( Conv1(x) ) ) ) ) + shortcut(x) )
        """
        # 主路径第一段：卷积 → BN → ReLU
        out = F.relu(self.bn1(self.conv1(x)))
        
        # 主路径第二段：卷积 → BN（注意：这里还没有ReLU！）
        out = self.bn2(self.conv2(out))
        
        # 残差连接：将主路径输出与捷径输出相加（核心操作！）
        # 这个加法让梯度可以"绕过"两层的卷积直接回传，解决了梯度消失问题
        out += self.shortcut(x)
        
        # 最后再通过ReLU激活（先加后激活）
        out = F.relu(out)
        return out


# ==================== ResNet：完整的残差网络 ====================
class ResNet(nn.Module):
    def __init__(self, block, num_blocks, num_classes=10):
        """
        参数：
            block: 残差块类型（BasicBlock 或 Bottleneck）
            num_blocks: 列表，每个阶段堆叠的残差块数量
                        例如 [2,2,2,2] 表示每个阶段2个块 → ResNet18
                        例如 [3,4,6,3] 表示四个阶段分别为3,4,6,3个块 → ResNet34
            num_classes: 分类类别数（CIFAR-10是10）
        """
        super(ResNet, self).__init__()
        
        # in_planes 记录当前层的输入通道数，会随着网络加深而更新
        self.in_planes = 64
        
        # ========== 初始卷积层（Stem）==========
        # 针对CIFAR-10（32x32图片）设计的初始层
        # 原版ResNet（针对ImageNet 224x224）使用7x7卷积 + 最大池化
        # 这里简化为单个3x3卷积，因为CIFAR-10图片更小
        self.conv1 = nn.Conv2d(3, 64, kernel_size=3, stride=1, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(64)
        
        # ========== 四个残差阶段（Layer1 ~ Layer4）==========
        # 每个阶段通过 _make_layer 堆叠多个残差块
        # 层数变化：
        #   layer1: 64 → 64（stride=1，不改变尺寸）
        #   layer2: 64 → 128（stride=2，尺寸减半，通道翻倍）
        #   layer3: 128 → 256（stride=2，尺寸减半，通道翻倍）
        #   layer4: 256 → 512（stride=2，尺寸减半，通道翻倍）
        # 最终特征图尺寸：32 → 16 → 8 → 4 → 4（经过全局池化后）
        self.layer1 = self._make_layer(block, 64, num_blocks[0], stride=1)
        self.layer2 = self._make_layer(block, 128, num_blocks[1], stride=2)
        self.layer3 = self._make_layer(block, 256, num_blocks[2], stride=2)
        self.layer4 = self._make_layer(block, 512, num_blocks[3], stride=2)
        
        # ========== 分类头 ==========
        # 全局平均池化 + 全连接层
        # 注意：原版ResNet在Global Avg Pooling后使用512 * block.expansion维度的全连接
        self.linear = nn.Linear(512 * block.expansion, num_classes)
    
    def _make_layer(self, block, planes, num_blocks, stride):
        """
        构建一个残差阶段（多个连续的残差块）
        
        参数：
            block: 残差块类型
            planes: 该阶段输出通道数
            num_blocks: 该阶段包含的残差块数量
            stride: 该阶段第一个残差块的步长（用于下采样）
        
        原理：
            - 第一个残差块：可能改变通道数和尺寸（当stride=2或通道数变化时）
            - 后续残差块：stride=1，保持通道数和尺寸不变
        """
        strides = [stride] + [1] * (num_blocks - 1)
        # 例如 num_blocks=2, stride=2 时：
        # strides = [2, 1]
        # 第1个块：stride=2（尺寸减半）
        # 第2个块：stride=1（尺寸不变）
        
        layers = []
        for stride in strides:
            # 每个块使用当前输入通道数 self.in_planes
            layers.append(block(self.in_planes, planes, stride))
            # 更新输入通道数为该块的输出通道数
            # BasicBlock 中 expansion=1，所以输出通道数 = planes
            self.in_planes = planes * block.expansion
        return nn.Sequential(*layers)
    
    def forward(self, x):
        """
        前向传播流程：
        输入(3x32x32) 
          → conv1 + bn1 + relu → (64x32x32)
          → layer1 (2个BasicBlock) → (64x32x32)
          → layer2 (2个BasicBlock) → (128x16x16)   # 第一次下采样
          → layer3 (2个BasicBlock) → (256x8x8)     # 第二次下采样
          → layer4 (2个BasicBlock) → (512x4x4)     # 第三次下采样
          → 全局平均池化 → (512x1x1) → flatten → (512)
          → 全连接层 → (10)
        """
        # 初始卷积 + BN + ReLU
        out = F.relu(self.bn1(self.conv1(x)))
        
        # 四个残差阶段
        out = self.layer1(out)
        out = self.layer2(out)
        out = self.layer3(out)
        out = self.layer4(out)
        
        # 全局平均池化（Global Average Pooling）
        # 作用：将每个特征图（512个）压缩成一个数值
        # 参数 kernel_size=4 是因为此时特征图尺寸是 4x4
        # 效果：512x4x4 → 512x1x1
        out = F.avg_pool2d(out, 4)
        
        # 展平：512x1x1 → 512
        out = out.view(out.size(0), -1)
        
        # 全连接层输出分类结果（10个类别）
        out = self.linear(out)
        return out


def ResNet18():
    """
    创建 ResNet-18 模型
    结构：初始卷积 + 4个阶段，每个阶段2个BasicBlock
    总层数：1 + (2+2+2+2)×2 = 17个卷积层 + 1个全连接层 = 18层
    """
    return ResNet(BasicBlock, [2, 2, 2, 2])


def ResNet34():
    """
    创建 ResNet-34 模型
    结构：初始卷积 + 4个阶段，阶段块数分别为[3,4,6,3]
    总层数：1 + (3+4+6+3)×2 = 33个卷积层 + 1个全连接层 = 34层
    """
    return ResNet(BasicBlock, [3, 4, 6, 3])
```

# 对CIFAR10训练

### 数据维度必须与网络第一层和最后一层匹配

```python
CIFAR-10 数据: [batch, 3, 32, 32]
                    ↓
           必须匹配网络第一层的输入
                    ↓
ResNet18 第一层: nn.Conv2d(3, 64, kernel_size=3, stride=1, padding=1)
                          ↑
                    这个 3 就是匹配图片的 3 通道
                    ↓
经过 ResNet18 各层后
                    ↓
ResNet18 最后一层: nn.Linear(512, 10)
                              ↑
                        这个 10 就是匹配 CIFAR-10 的 10 个类别
```

```python
from torch.utils.data import DataLoader
from torch.utils.tensorboard import SummaryWriter


# ==================== 准备数据集 ====================
train_data = torchvision.datasets.CIFAR10(root="data", train=True, 
                                          transform=torchvision.transforms.ToTensor(), 
                                          download=True)
test_data = torchvision.datasets.CIFAR10(root="data", train=False, 
                                         transform=torchvision.transforms.ToTensor(), 
                                         download=True)

train_data_size = len(train_data)
test_data_size = len(test_data)
print(f"训练数据集的长度为：{train_data_size}")
print(f"测试数据集的长度为：{test_data_size}")

# DataLoader
train_dataloader = DataLoader(train_data, batch_size=64, shuffle=True)
test_dataloader = DataLoader(test_data, batch_size=64, shuffle=False)

# ==================== 创建模型 ====================
net = ResNet18()  # 这里换成 ResNet18

# 损失函数和优化器
loss_fn = nn.CrossEntropyLoss()
learning_rate = 0.001
optimizer = torch.optim.SGD(net.parameters(), lr=learning_rate)

# 训练参数
total_train_step = 0
total_test_step = 0
epoch = 10

# TensorBoard
writer = SummaryWriter("logs_train_resnet18")  # 换个名字，避免和之前的混淆

# ==================== 开始训练 ====================
for i in range(epoch):
    print(f"第{i+1}轮训练开始")
    
    # 训练步骤
    net.train()
    for data in train_dataloader:
        imgs, targets = data
        outputs = net(imgs)
        loss = loss_fn(outputs, targets)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        total_train_step += 1
        if total_train_step % 100 == 0:
            print(f"训练次数：{total_train_step}, 损失值：{loss.item():.4f}")
            writer.add_scalar("train_loss", loss.item(), total_train_step)
    
    # 测试步骤
    net.eval()
    total_test_loss = 0
    total_accuracy = 0
    
    with torch.no_grad():
        for data in test_dataloader:
            imgs, targets = data
            outputs = net(imgs)
            
            loss = loss_fn(outputs, targets)
            total_test_loss += loss.item()
            
            predictions = outputs.argmax(dim=1)
            correct = (predictions == targets).sum().item()
            total_accuracy += correct
    
    avg_test_loss = total_test_loss / len(test_dataloader)
    test_accuracy = total_accuracy / test_data_size
    
    print(f"第{i+1}轮测试集 - Loss: {avg_test_loss:.4f}, 准确率: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")
    
    writer.add_scalar("test_loss", avg_test_loss, total_test_step)
    writer.add_scalar("test_accuracy", test_accuracy, total_test_step)
    total_test_step += 1
    
    # 保存模型
    torch.save(net.state_dict(), f"resnet18_epoch_{i}.pth")
    print(f"模型已保存为 resnet18_epoch_{i}.pth")

writer.close()
print("训练完成！")
```

# 模型验证

```python
from PIL import Image
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision

# ==================== 定义 ResNet18 模型 ====================
class BasicBlock(nn.Module):
    expansion = 1
    def __init__(self, in_planes, planes, stride=1):
        super(BasicBlock, self).__init__()
        self.conv1 = nn.Conv2d(in_planes, planes, kernel_size=3, stride=stride, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(planes)
        self.conv2 = nn.Conv2d(planes, planes, kernel_size=3, stride=1, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(planes)

        self.shortcut = nn.Sequential()
        if stride != 1 or in_planes != self.expansion * planes:
            self.shortcut = nn.Sequential(
                nn.Conv2d(in_planes, self.expansion * planes, kernel_size=1, stride=stride, bias=False),
                nn.BatchNorm2d(self.expansion * planes)
            )

    def forward(self, x):
        out = F.relu(self.bn1(self.conv1(x)))
        out = self.bn2(self.conv2(out))
        out += self.shortcut(x)
        out = F.relu(out)
        return out

class ResNet(nn.Module):
    def __init__(self, block, num_blocks, num_classes=10):
        super(ResNet, self).__init__()
        self.in_planes = 64
        self.conv1 = nn.Conv2d(3, 64, kernel_size=3, stride=1, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(64)
        self.layer1 = self._make_layer(block, 64, num_blocks[0], stride=1)
        self.layer2 = self._make_layer(block, 128, num_blocks[1], stride=2)
        self.layer3 = self._make_layer(block, 256, num_blocks[2], stride=2)
        self.layer4 = self._make_layer(block, 512, num_blocks[3], stride=2)
        self.linear = nn.Linear(512 * block.expansion, num_classes)

    def _make_layer(self, block, planes, num_blocks, stride):
        strides = [stride] + [1] * (num_blocks - 1)
        layers = []
        for stride in strides:
            layers.append(block(self.in_planes, planes, stride))
            self.in_planes = planes * block.expansion
        return nn.Sequential(*layers)

    def forward(self, x):
        out = F.relu(self.bn1(self.conv1(x)))
        out = self.layer1(out)
        out = self.layer2(out)
        out = self.layer3(out)
        out = self.layer4(out)
        out = F.avg_pool2d(out, 4)
        out = out.view(out.size(0), -1)
        out = self.linear(out)
        return out

def ResNet18():
    return ResNet(BasicBlock, [2, 2, 2, 2])


# ==================== 验证代码 ====================

# 1. 加载并预处理图片
image_path = "./imgs/cat_buding.png"  # 改成你的图片路径
image = Image.open(image_path)
print(f"原始图片: {image}")

# 图片预处理（和训练时保持一致）
transform = torchvision.transforms.Compose([
    torchvision.transforms.Resize((32, 32)),  # ResNet18 输入也是 32x32
    torchvision.transforms.ToTensor()
])

image = transform(image)
print(f"预处理后形状: {image.shape}")  # torch.Size([3, 32, 32])

# 2. 加载模型
model = ResNet18()  # 创建模型实例
state_dict = torch.load("net_epoch_25.pth", map_location=torch.device('cpu'), weights_only=False)
model.load_state_dict(state_dict)
print("模型加载成功！")

# 3. 打印模型结构（可选）
print(model)

# 4. 预测
image = torch.reshape(image, (1, 3, 32, 32))  # 添加 batch 维度
model.eval()  # 转为评估模式
with torch.no_grad():  # 节省内存
    output = model(image)

print(f"原始输出: {output}")
print(f"预测的类别索引: {output.argmax(1).item()}")

# 5. CIFAR-10 类别名称（可选，显示具体类别）
classes = ['airplane', 'automobile', 'bird', 'cat', 'deer', 
           'dog', 'frog', 'horse', 'ship', 'truck']
pred_class = classes[output.argmax(1).item()]
print(f"预测的类别: {pred_class}")
```
